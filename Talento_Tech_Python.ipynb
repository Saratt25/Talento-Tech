```python
#Realizar un algoritmo que calcule el promedio de la edad de n personas
n = int(input("Ingrese el numero de personas: "))   #declaramos variable n para personas
k = 1   #definimos variable para CONTROLAR INTERACCIONES
promedio = 0.0   #definimos variable promedio para almacenar valor promedio
while k <= n:  #establecemos condicion para ejecucion del ciclo
  edad = int(input("Ingresar edad"))   #ingresamos edades
  promedio = promedio + edad   #acumulamos las edades
  k = k + 1   #actualizo el contador
print(promedio/n)   #calculo e imprimo mi promedio
```

```python
#Realizar un algoritmo para comparar las areas de un triangulo y un cuadrado
#...al final debe indicar cual de ellas es mayor
tb = float(input("ingrese la base del triangulo:"))
ta = float(input("ingrese la altura del triangulo:"))
areaT = (tb*ta)/2
ca = float(input("ingrese el lado del cuadrado:"))
areaC = ca*ca
if areaT > areaC:
  print("El area del triangulo es mayor")
else:
  if areaT < areaC:
    print("El area del cuadrado es mayor")
  else:
      print("Los areas son iguales")


```

```python
i = 1 #controlar altura
j = 1 #controlar ancho
n = int(input("tamaño de triangulo:"))
while i <= n:  #llenado o recorrido de la altura del triangulo
  j = 1
  while j <= i:  #llenado del ancho del triangulo
    print("*", end="")  #imprimiendo el caracter, pero solo uno, sin salto de linea
    j = j + 1
  print("")  #salto de linea
  i = i + 1
```

```python
#triangulo invertido
i = 1 #controlar altura
n = int(input("tamaño de triangulo:"))
while i <= n:  #llenado o recorrido de la altura del triangulo
  j = 0
  while j <= n-i:  #llenado del ancho del triangulo
    print("*", end="")  #imprimiendo el caracter, pero solo uno, sin salto de linea
    j = j + 1
  print("")  #salto de linea
  i = i + 1
```
